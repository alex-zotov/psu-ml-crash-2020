### 1 Подсчёт уникальных символов в строке, возвращает словарь

#### может можно как-то более pythonic way?

```python
def count_unique(str):
    d={}
    for ch in str:
        if ch in d:
            d[ch]+=1
        else:
            d[ch]=1
    return d    
```

```python
print(count_unique('uniqu§e strng'))
```

### 2 Чтение csv

```python
import re
```

```python
def read_csv(file):
    p=re.compile(r'(\,|\r?\n|\r|^)(?:"([^"]*(?:""[^"]*)*)"|([^"\,\r\n]*))')
    with open(file) as f:
        title,data=[],[]
        for line in f:
            #празбиваем строку по разделителям
            pf=p.findall(line.rstrip())
            #заменяем двойные кавычки на одинарные, там где было экранирование - это группа патерна p1
            col=[p1.replace('""','"') if p1 else p2 for p0,p1,p2 in pf]
                
            if title:
                data.append(col)
            else:
                title=col
    return title,data
#проверим что длина строки с данными та же что и умассива с названем колонок
#если где-то требование нарушено вернем индекс строки с данными
def err_len(df):
    cnt=len(df[0])
    err=-1
    for i,r in enumerate(df[1]):
        if len(r)!=cnt:
            err=i
            break
    return err
```

```python
df0=read_csv('titanic.csv')
#проверим, что не ошиблись с длиной строк
print(err_len(df0))
#название колонок
print(df0[0])
#первые 10 строк даных
print(df0[1][:10])
```
